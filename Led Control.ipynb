### Led Control - Finger Detection

```python
import cv2
import mediapipe as mp
import time
import Controller as cnt

time.sleep(2.0)

video = cv2.VideoCapture(0)         # 0 for laptop camera    # 1 for webcam

mpHands = mp.solutions.hands
#hands = mpHands.Hands()
mpDraw = mp.solutions.drawing_utils

tipIds = [4, 8, 12, 16, 20]

with mpHands.Hands(max_num_hands = 1, min_detection_confidence = 0.8,
                   min_tracking_confidence = 0.8) as hands:

    while True:
        success, img = video.read()
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        img.flags.writeable = False
        results = hands.process(img)
        img.flags.writeable = True
        img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)

        lmlist = []
        
        if results.multi_hand_landmarks:
            for handLms in results.multi_hand_landmarks:
                myHands1 = results.multi_hand_landmarks[0]
                for id, lm in enumerate(myHands1.landmark):
                    h, w, c = img.shape
                    cx, cy = int(lm.x*w), int(lm.y*h)
                    #lmlist.append([id, cx, cy])
                    lmlist.append([id, lm.x, lm.y])
                mpDraw.draw_landmarks(img, handLms, mpHands.HAND_CONNECTIONS)
                
        
        fingers = []
        if len(lmlist) != 0:
            if lmlist[tipIds[0]][1] > lmlist[tipIds[0]-1][1]:
                fingers.append(1)
            else:
                fingers.append(0)
            
            for id in range (1,5):
                if lmlist[tipIds[id]][2] < lmlist[tipIds[id]-2][2]:
                    fingers.append(1)
                else:
                    fingers.append(0)
        
        total = fingers.count(1)
        cnt.led(total)
        
        if total == 0:
            
            cv2.rectangle(img, (20, 300), (400, 425), (0, 255, 0), 
                          cv2.FILLED)
            cv2.putText(img, "0", (45, 375), cv2.FONT_HERSHEY_COMPLEX, 
                        3, (255, 0, 255), 3)
            cv2.putText(img, " LED", (100, 375), cv2.FONT_HERSHEY_COMPLEX, 
                        3, (255, 0, 255), 3)
        if total == 1:
            
            cv2.rectangle(img, (20, 300), (400, 425), (0, 255, 0), 
                          cv2.FILLED)
            cv2.putText(img, "1", (45, 375), cv2.FONT_HERSHEY_COMPLEX, 
                        3, (255, 0, 255), 3)
            cv2.putText(img, " LED", (100, 375), cv2.FONT_HERSHEY_COMPLEX, 
                        3, (255, 0, 255), 3)
        if total == 2:
            
            cv2.rectangle(img, (20, 300), (400, 425), (0, 255, 0), 
                          cv2.FILLED)
            cv2.putText(img, "2", (45, 375), cv2.FONT_HERSHEY_COMPLEX, 
                        3, (255, 0, 255), 3)
            cv2.putText(img, " LEDS", (100, 375), cv2.FONT_HERSHEY_COMPLEX, 
                        3, (255, 0, 255), 3)
        if total == 3:
            
            cv2.rectangle(img, (20, 300), (400, 425), (0, 255, 0), 
                          cv2.FILLED)
            cv2.putText(img, "3", (45, 375), cv2.FONT_HERSHEY_COMPLEX, 
                        3, (255, 0, 255), 3)
            cv2.putText(img, " LEDS", (100, 375), cv2.FONT_HERSHEY_COMPLEX, 
                        3, (255, 0, 255), 3)
        if total == 4:
            
            cv2.rectangle(img, (20, 300), (400, 425), (0, 255, 0), 
                          cv2.FILLED)
            cv2.putText(img, "4", (45, 375), cv2.FONT_HERSHEY_COMPLEX, 
                        3, (255, 0, 255), 3)
            cv2.putText(img, " LEDS", (100, 375), cv2.FONT_HERSHEY_COMPLEX, 
                        3, (255, 0, 255), 3)
        if total == 5:
            
            cv2.rectangle(img, (20, 300), (400, 425), (0, 255, 0), 
                          cv2.FILLED)
            cv2.putText(img, "5", (45, 375), cv2.FONT_HERSHEY_COMPLEX, 
                        3, (255, 0, 255), 3)
            cv2.putText(img, " LEDS", (100, 375), cv2.FONT_HERSHEY_COMPLEX, 
                        3, (255, 0, 255), 3)
            
            
        cv2.imshow('Image', img)
        k = cv2.waitKey(1)
        if k == ord('q'):
            break
        
video.release()
cv2.destroyAllWindows()
```
